#### Total Electricity generated per Country

```python
# group the data by country and sum the electricity generation values
grouped = ElectricityGenerated_EU.groupby('Entity')['Electricity generation (TWh)'].mean()
grouped_sorted = grouped.sort_values(ascending=True)
print(grouped_sorted)
```

#### Visualizing the evolution of electricity generated for each country

```python
# group the dataframe by entity
grouped = ElectricityGenerated_EU.groupby('Entity')

fig, ax = plt.subplots(figsize=(28, 12))
for name, group in grouped:
    ax.plot(group['Year'], group['Electricity generation (TWh)'], label=name)

# set the title and labels for the plot
ax.set_title('Evolution of electricity generation with year per entity')
ax.set_xlabel('Year')
ax.set_ylabel('Electricity generation (TWh)')

# add a legend to the plot
ax.legend()

# show the plot
plt.show()
```

## Exploratory Analysis and visualization 

### Exploration of features (unique)

```python
SourceOfElectricityProduction.nunique()
# this gives us the number of unique values. 
```

```python
SourceOfElectricityProduction = SourceOfElectricityProduction.loc[SourceOfElectricityProduction['Entity'].isin(["Austria","Belgium","Bulgaria",
"Croatia","Denmark", "Estonia", "Finland", "France", "Germany", "Greece", "Hungary", "Ireland", "Italy","Latvia", "Lithuania", 
"Luxembourg", "Malta", "Netherlands", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Spain" , "Sweden","Cyprus","Czechia"])]
```

### visualize cartegorical and continues data to see trends.

#### Top 5 countries with the highest electricity generation from each source visaulization

```python
# Create a dictionary to store the top 5 countries for each source
top_countries = {}

# Loop through each electricity source and find the top 5 countries
for source in ['Other renewables excluding bioenergy (TWh) (zero filled)', 'Electricity from bioenergy (TWh) (zero filled)', 'Electricity from solar (TWh)', 'Electricity from wind (TWh)', 'Electricity from hydro (TWh)', 'Electricity from nuclear (TWh)', 'Electricity from oil (TWh)', 'Electricity from gas (TWh)', 'Electricity from coal (TWh)']:
    top_countries[source] = SourceOfElectricityProduction.groupby('Entity')[source].sum().nlargest(5).index.tolist()

# Create subplots for each electricity source
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))

# Loop through each electricity source and create a bar chart for the top 5 countries
for i, source in enumerate(top_countries.keys()):
    ax = axes[i//3][i%3]
    df_top = SourceOfElectricityProduction[SourceOfElectricityProduction['Entity'].isin(top_countries[source])]
    df_top = df_top.groupby('Entity')[source].sum().reset_index()
    df_top = df_top.sort_values(source, ascending=False)
    ax.bar(df_top['Entity'], df_top[source])
    ax.set_title(source)

# Adjust the spacing between subplots
plt.subplots_adjust(hspace=0.4)

# Show the plot
plt.show()
```

#### correlation between different electricity sources

```python
columns = ['Other renewables excluding bioenergy (TWh) (zero filled)', 'Electricity from bioenergy (TWh) (zero filled)', 'Electricity from solar (TWh)', 'Electricity from wind (TWh)', 'Electricity from hydro (TWh)', 'Electricity from nuclear (TWh)', 'Electricity from oil (TWh)', 'Electricity from gas (TWh)', 'Electricity from coal (TWh)']
data = SourceOfElectricityProduction[columns]

# Calculate the correlation matrix
corr_matrix = data.corr()

# Print the correlation matrix
print(corr_matrix)
```
